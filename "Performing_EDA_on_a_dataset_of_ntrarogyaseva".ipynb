#### **Exploratory Data Analysis**
Exploratory Data Analysis (EDA) is the process of examining and summarizing a dataset to understand its characteristics, identify patterns, and make informed decisions. It involves calculating summary statistics, visualizing data through plots and charts, identify missing or incosistent data, and exploring relationships between variables. EDA provides insights that guide further analysis and decision-making.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(color_codes=True)
```

### Finding out and Dropping the missing or null values

```python
print(df.isnull().sum())
```

### Data Exploration

```python
# Display the counts of each value in the sex column
df['SEX'].value_counts()
```

It appears that there are duplicate values ​​in this column. Male and MALE are not two different sexes. We can substitute the column names to resolve this issue. I will also rename Male (Child) -> Boy and Female (Child) -> Girl for convenience:

```python
# Mappings to standardize and clean the values\
mappings = {'MALE':'Male', 'FEMALE':'Female','Male(Child)':'Boy','Female(Child)':'Girl'}
# Replace values using the defined mappings
df['SEX'] = df['SEX'].replace(mappings)
df['SEX'].value_counts()
```

```python
# Plot the Value counts of Sex
df['SEX'].value_counts().plot.bar()
```

**Age Distribution of the Data.**
###### Mean, Median and Mode of the Data

```python
#Print the Mean, Median and mode of the age Distribution
print("Mean: {}".format(df['AGE'].mean()))
print("Median: {}".format(df['AGE'].median()))
print("Mode: {}".format(df['AGE'].mode()))
```

```python
# Print the top 10 pages
df['AGE'].value_counts().head(10)
```

```python
# Boxplot for age variable
df['AGE'].plot.box()
```

```python
# Boxplot (using seaborn)  for age variable
sns.boxplot(x=df['AGE'])
```

**Analysing only records pertaining to the district of Krishna.**

```python
# Subset involving only records of Krishna district
df[df['DISTRICT_NAME']=='Krishna'].head()
```

Most prevalent surgery, district wise

```python
# Most common surgery by district
for i in df['DISTRICT_NAME'].unique():
  print("District: {}nDisease and Count: {}".format(i,df[df['DISTRICT_NAME']==i]['SURGERY'].value_counts().head(1)))
```

We can observe that only two surgeries top all the districts:

* Dialysis (7 districts)
* Longbone Fracture (6 districts)

**Average claim amount, district wise**

```python
# Average claim amount for surgery by district
for i in df['DISTRICT_NAME'].unique():
    print("District: {}nAverage Claim Amount: ₹{}".format(i,df[df['DISTRICT_NAME']==i]['CLAIM_AMOUNT'].mean()))
```

**Statistics by surgery category**

```python
#Group by surgery category to get mean statistics 
df.groupby('CATEGORY_NAME').mean()
```

**Most common surgery by age group**

```python
# Create a new memory copy of data to manipulate age
dataround = df.copy()
```

```python
dataround['AGE'] = dataround['AGE'].round(-1)
```

Visualizing the age groups using seaborn's countplot function

```python
# A frequency plot for each age group
sns.countplot(x=dataround['AGE'])
```

**Most common surgery per age group**

```python
# Most common surgery and count per age group
for i in sorted(dataround['AGE'].unique()):
  print("Age Group: {}\nMost Common Surgery and Count: {}".format(i,df[df['AGE']==i]['CATEGORY_NAME'].value_counts().head(1)))
```
